# Imports

```python
import pandas as pd
```

```python
data = pd.read_csv("Titanic-Dataset.csv")
# print(data)
```

```python
df = pd.DataFrame(data)
```

```python
w_missing = pd.isnull(df)
print(w_missing.sum())
```

# Dropping Embarked

```python
missing = pd.isnull(df["Embarked"])
print(missing.sum()) #number of missing values for embarked now 0
df =  df[~missing]
print(df)
```

# Processing Sex

```python
df["Sex"] = df["Sex"].replace({"male":0, "female":1})
print(df)
```

# Process Embarked using one-hot

```python
df["Embarked"] = df["Embarked"].replace({"C":"100", "Q":"010", "S":"001"})
print(df)
```
